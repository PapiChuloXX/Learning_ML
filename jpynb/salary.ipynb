```python
from csv import DictReader
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
output_label = []
feature_set = []

with open("Salary_dataset.csv") as f:
    file = DictReader(f)
    for row in file:
        output_label.append(float(row['Salary']))
        feature_set.append(float(row['YearsExperience']))
```

```python
f = lambda b1,x: b1*(x) + min(output_label)

def cost_function(slope_vector, x_vector, y_vector):
    cost = [f(w,x)for w,x,y in zip(slope_vector, x_vector, y_vector)]
    return cost

y_estimate = [f(7763.3,x) for x in feature_set]
plt.cla()
plt.plot(feature_set,y_estimate,"ro")
plt.plot(feature_set,y_estimate)
plt.plot(feature_set,output_label, "bo")
plt.show()
```

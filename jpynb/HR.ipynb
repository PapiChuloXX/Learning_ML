```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error
```

```python
training_data =  pd.read_csv("HR_comma_sep.csv")
training_data.describe()
```

```python
training_features = ["satisfaction_level", "last_evaluation", 
                        "number_project", "average_montly_hours",
                            "time_spend_company", "Work_accident", "promotion_last_5years"]

X_train, X_test, y_train, y_test = train_test_split(training_data[training_features], 
                                                        training_data[["left"]],
                                                            random_state=50, train_size=0.7)

X_train.describe()
```

```python
model = Sequential(
        [Dense(input_dim = X_train.shape[1],units=100,activation="relu"),
         Dense(units=25,activation="relu"),
         Dense(units=10,activation="relu"),
         Dense(units=1, activation="sigmoid")
        ], name = "my_model")

model.summary()
```

```python
model.compile(optimizer = 'adam', loss="binary_crossentropy")
model.fit(X_train,y_train,epochs=1000)
```

```python
predictions = model.predict(X_test)

threshold = 0.5

predictions = np.where(predictions < threshold, np.floor(predictions), np.ceil(predictions))
print(1 - mean_absolute_error(predictions, y_test))
```

```python
model.evaluate(X_test, y_test)
```

```python
z = lambda x,w,b: np.dot(w,x) + b
```

```python
g =  lambda z: 1 / (1 + np.exp(-z))
```

```python
def compute_gradient(X,y,w,b):
    dj_dw = np.zeros(w.shape)
    dj_db = 0.
    m = X.shape[0]
    X, y = X.values, y.values

    for i in range(m):
        loss = g(z(X[i],w,b)) - y[i]
        dj_dw += loss * X[i]
        dj_db += loss[0]

    return dj_dw / m , dj_db / m
```

```python
def gradient_descent(X,y,w,b,nums_iter,alpha):
    for i in range(nums_iter):
        dj_dw, dj_db = compute_gradient(X,y,w,b)
        w -= alpha * dj_dw
        b -= alpha * dj_db

    return w,b
```

```python
w_,b_ = gradient_descent(X_train,y_train,np.zeros(7),0.,1000,0.01)
```

```python
def predict(X):
    m = X.shape[0]
    predictions = np.zeros(m)
    X = X.values

    for i in range(m):
        res = g(z(X[i], w_, b_))
        predictions[i] = np.ceil(res) if res >= 0.5 else np.floor(res)

    return predictions


y_pred = predict(X_test)
print(1 - mean_absolute_error(y_test, y_pred))
```

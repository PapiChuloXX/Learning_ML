```python
import numpy as np
import pandas as pd
import math
%matplotlib widget
import matplotlib.pyplot as plt
import csv
```

```python
x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([1.0, 2.0, 3.0])
b1 = 0
f = lambda input, b0 :b0*input + b1


def J(slope,features,outcomes):
    cost = sum([(f(feature,slope) - outcome)**2 for feature,outcome in zip(features, outcomes)]) / (2*features.shape[0])
    return cost

w_vector = [-0.5,0,0.5,1.0,1.5,2.0,2.5]
cost = [J(w,x_train,y_train) for w in w_vector]
print(cost)
```

```python
plt.cla()
plt.plot(w_vector, cost, "ro")
plt.plot(w_vector, cost)
plt.show()
```
